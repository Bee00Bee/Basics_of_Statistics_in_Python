## Data Visualization: Numeric data

In this notebook we want to explore the data visualization. There are definitions here which we need to know:
* Boxplot: A plot which presents the locality, skewness of the data. The box plot also presents these imporant measured values: Minimum, First Quartile, median, Third Quartile, and Maximum. 
* Frequency Table: The count of data points in the dataset.
* Histogram: Is the plot of the frequency table with specified bins.
* Density Plot: It presents the distribution of numeric values.

```python
from numpy.random import seed
from numpy.random import randn
import matplotlib.pyplot as plt
import numpy as np
import scipy
from numpy import mean
from numpy import std
```

### Box Plot

```python
data = 23 * randn(200) + 40
```

```python
plt.boxplot(data)
plt.show()
```

In the box plot the median is shown by a red line which is around 50. The bottom and top part of the box are the 25th and 75th percentiles, respectively. Between the 25th percentile and 75th percentile is the interquartile range. The whiskers shown in a black like which are extending on the both sides of the box and outliers are the circles above and below the whiskers. The horizontal lines which are located between the whiskers and the outliers are the boundaries which can define the outliers.

### Density Plot

```python
import seaborn as sns
sns.distplot(data, color = 'purple')
```

As we can see from the density plot the average of the data lies around 50 and the majority of the data is between 0 and 75 and the densed part of the data is in (25,70) range. This can be seen from the histogram plot below. 

### Histogram Plot

```python
plt.hist(data, color = 'purple')
plt.show()
```

## Data Visualization: Binary and Categorical Data
* Bar chart: The frequency of data points for each data point is presented as bar chart.
* Pie chart: The frequency of data points is plotted as slices in a pie chart.

### Bar Chart

```python
data = ([1,3,4,5,6,9],[3,4,1,5,10,6])
print("The data points are: ", data[0],"\nthe frequency of the data is: ",data[1])
plt.bar(data[0],data[1])
plt.show()
```

The most frequent data is 6 then 9 and the third one is 5. 

### Pie Chart

```python
data = np.array([3,4,1,5,10,6])
mylabels = ["1","3","4","5","6","9"]

plt.pie(data, labels = mylabels)
plt.show()
```

Here the number 6 is repeated the most then number 9 and number 5 in the dataset. 

### Mode
The mode presents the most frequent data point in the dataset. Let's calculate it:

```python
data = [6,6,6,6,6,6,6,6,6,6,9,9,9,9,9,9,5,5,5,5,5,4,3,3,3,3,1,1,1]
```

```python
mode = scipy.stats.mode(data)
print("The mode of the data is number: ", mode[0][0])
```

Here I just want to add one more plot which is *scatter plot* which presents the reslationship between two vairables. Let's say we want to see the relationship between the data points and the frequency of the data.

```python
data2 = ([1,3,4,5,6,9],[3,4,1,5,10,6])
plt.plot(data2[0], data2[1],'-o')
plt.xlabel('The numbers')
plt.ylabel('Frequency of the each data point')
```
